**Nested array**: Arrays that have arrays as their elements

**Dimension**: Dimension in arrays is one level of array depth.

**0-D array**:
- e.g. 2, 45
- also known as scalars

**1-D array**:
- [1, 2, 3, 4, 5]
- uni-dimensional array
- has 0-D array as its elements.

**2-D array**:
- [[1, 2, 3], [2, 5, 7]]
- Bi-dimensional array
- has 1-D array as its elements

**3-D array**:
- [ [[1, 2, 3], [2, 5, 7]], [[1, 2, 5], [3, 5, 9]] ]
- has 2-D array as its elements
- (2, 2, 3)

```python
import numpy as np

class Linear:
    def __init__(self, input_dim, output_dim):
        # Initialize weights randomly and biases to zero
        # We divide by sqrt(input_dim) to keep starting variance small (Xavier/He initialization)
        self.W = np.random.randn(input_dim, output_dim) / np.sqrt(input_dim)
        self.b = np.zeros(output_dim)

        # We will store X here during the forward pass so we can use it in the backward pass
        self.X_cache = None

        # Gradients will be stored here
        self.dW = None
        self.db = None

    def forward(self, X):
        """
        X shape: (batch_size, input_dim)
        Returns: Z shape (batch_size, output_dim)
        """
        # Save X for the backward pass
        self.X_cache = X

        # TODO: Calculate Z = XW + b
        Z = X @ self.W + self.b

        return Z

    def backward(self, dZ):
        """
        dZ shape: (batch_size, output_dim)
        Returns: dX shape (batch_size, input_dim)
        """
        X = self.X_cache

        # TODO: Calculate dW, db, and dX based on the formulas above
        self.dW = X.T @ dZ
        self.db = np.sum(dZ, axis=0)
        dX = dZ @ self.W.T

        return dX
```

```python
# stable_softmax()

def naive_softmax(X):
  exps = np.exp(X)
  return exps/np.sum(exps, axis=-1, keepdims=True)


```

```python
naive_softmax([1000, 1001, 1002])
```

```python
def stable_softmax(X):
  exps = np.exp(X - np.max(X, axis=-1, keepdims=True))
  return exps/np.sum(exps, axis=-1, keepdims=True)
```

```python
stable_softmax([1000, 1001, 1002])
```
